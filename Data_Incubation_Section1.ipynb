```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# The numbers are in percent except for average hourly wage which is in $
# Disclose_pct: Percentage of employers aware that young adults have a particular disability
# Employers_same_disability: Percentage of Employers that have the same disability as the young adult
```

```python
Disability_Table = pd.read_csv('Data_Files/Data_Incubator_Section1.csv')
Disability_Table
```

```python
df = Disability_Table[['disability_identifier','promotion', 'avg_months_retention']]

df.set_index('disability_identifier')
df.plot(kind = 'barh', x = 'disability_identifier', rot = 0)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05))
```

```python
df2 = Disability_Table[['disability_identifier', 'disclose_pct', 'employees_same_disability']]
df2.set_index('disability_identifier')
df2.plot(kind = 'barh', x = 'disability_identifier', rot = 0)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05))
```
